## Training a Machine Learning Model

We have explored the dataset, thoroughly analyzed it, determined what information we need, saved it into a usable format and finally, we are ready to build a machine learning model. For this task, we will be using [scikit-learn](https://scikit-learn.org/stable/).

The purpose of the model is to predict the pollution for the next 24 hours. 

```python
import os
import sys

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.metrics import explained_variance_score, mean_squared_error, mean_absolute_error
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import ExtraTreesRegressor


module_path = os.path.abspath(os.path.join(".."))
if module_path not in sys.path:
    sys.path.append(module_path)

from pipeline import model_selection
from pipeline.common import *

np.random.seed(0)

make_directory_tree(["plots", "models"], "../data/output")
```

```python
# read and describe the prepared dataset for Centar
df = read_csv_series("../data/centar.csv")
describe_series(df)
```

```python
# number of hours we need to predict ahead
predict_window = 12

# shift PM10 for `predict_window` hours ahead
df["PM10"] = df["PM10"].shift(-predict_window)

# split dataset into train, test and evaluation by dates
# additionally, leave the last hours for final evaluation
train_len = int(len(df) * 0.65) - (2 * predict_window)
test_len = int(len(df) * 0.25) - (2 * predict_window)
eval_len = len(df) - train_len - test_len - (2 * predict_window)
train, test, evaluation = df[:train_len], df[train_len:train_len +
                                                test_len], df[train_len+test_len:train_len+test_len+eval_len]
final_eval = df[-(2 * predict_window):-predict_window].copy()

# plot the splits
plt.figure(figsize=(15,6))
plt.xlabel('time')
plt.ylabel('PM10')
plt.plot(train.index, train["PM10"], label="train")
plt.plot(test.index, test["PM10"], label="test")
plt.plot(evaluation.index, evaluation["PM10"], label="eval")
plt.plot(final_eval.index, final_eval["PM10"], label="final")
plt.legend()
```

```python
output_dir = "../data/output"
lr = LinearRegression()
etr = ExtraTreesRegressor(n_estimators=50, random_state=0)

model_selection.train_and_evaluate("Linear Regression", lr, train, test, evaluation, final_eval, output_dir)
model_selection.train_and_evaluate("Extra Trees Regressor", etr, train, test, evaluation, final_eval, output_dir)
```

## Run Script For All Datasets

```python
model_selection.run("../data", output_dir)
```
