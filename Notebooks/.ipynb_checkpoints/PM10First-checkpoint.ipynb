```python
#replacing windBearing with 0 according to darksky api
train['windBearing'].fillna(0,inplace=True)
```

```python
# Should be probably imputed with the average median of that hour in that month of the year
unique_hours = list(set(list(map(lambda x: x[5:10], train['time'].values))))
pat = '|'.join(unique_hours)
s = train['time'].str.extract('('+ pat + ')', expand=False)
median_pressure_values = train.groupby(s)['pressure'].median()

def transform_pressure(row):
    if np.isnan(row['pressure']):
        row['pressure'] = median_pressure_values[row['time'][5:]]
    return row
    
train = train.apply(transform_pressure,axis=1)
```

```python
median_precip = train.groupby(s)['precipProbability'].median()

def transform_precipProb(row):
    if np.isnan(row['precipProbability']):
        row['precipProbability'] = median_precip[row['time'][5:]]
    return row
    
train = train.apply(transform_precipProb,axis=1)

median_precip_i = train.groupby(s)['precipIntensity'].median()

def transform_precipInt(row):
    if np.isnan(row['precipIntensity']):
        row['precipIntensity'] = median_precip_i[row['time'][5:]]
    return row
    
train = train.apply(transform_precipInt,axis=1)

median_uvI = train.groupby(s)['uvIndex'].median()

def transform_medianUvi(row):
    if np.isnan(row['uvIndex']):
        row['uvIndex'] = median_uvI[row['time'][5:]]
    return row
    
train = train.apply(transform_medianUvi,axis=1)

median_wSpeed = train.groupby(s)['windSpeed'].median()

def transform_windSpeed(row):
    if np.isnan(row['windSpeed']):
        row['windSpeed'] = median_wSpeed[row['time'][5:]]
    return row
    
train = train.apply(transform_windSpeed,axis=1)
```

```python
#Probably some of the medians had nans 
train['uvIndex'].fillna(train['uvIndex'].mean(),inplace=True)
```

```python
# drop columns that have more than 70% missing values
train.drop(['CO2','CO','SO2','NO2'], axis=1,inplace=True)
# We will generate new AQI 
train.drop(['AQI'], axis=1,inplace=True)
```

```python
#Feature engineer here, i toa viktor shto kazha za time series da se dolovi
# i sega tuka ako navlezesh povekje vo time series x(t+1) mozhe da rechish deka ti zavisit samo od x(t)
# ama mojsh da rechish deka ti zavisit i od x(t-1), x(t-2), x(t-3)
# taka da mozhda podobri rezultati bi dobil ako kako features gi dodadesh primer poslednite 5 merenja
# Vo princip, nekoj informed feature engineering + uninformed random feature engineering + time series feature

median_o3 = train.groupby(s)['O3'].median()

def transform_o3(row):
    if np.isnan(row['O3']):
        row['O3'] = median_o3[row['time'][5:]]
    return row
    
train = train.apply(transform_o3,axis=1)

median_pm25 = train.groupby(s)['PM25'].median()

def transform_pm25(row):
    if np.isnan(row['PM25']):
        row['PM25'] = median_pm25[row['time'][5:]]
    return row
    
train = train.apply(transform_pm25,axis=1)

#After clean
train["PM25"].interpolate(inplace=True, limit=3,method='time',limit_direction='both') # adapt limit here
train["O3"].interpolate(inplace=True, limit=3,method='time',limit_direction='both') # adapt limit here

train['PM25'].fillna(train['PM25'].mean(),inplace=True)
train['O3'].fillna(train['O3'].mean(),inplace=True)

total = train.isnull().sum().sort_values(ascending=False)
percent = (train.isnull().sum()/train.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data = missing_data[missing_data['Total']>0]
missing_data
```

```python
train.drop("apparentTemperature",axis=1,inplace=True) #high correlation
def mapFunckijaPM25(row):
    row['AQI_PM25'] = train[row.name - datetime.timedelta(days=1) : row.name]['PM25'].mean()
    return row
    #return train['PM25'][datetime.datetime[]]
train = train.apply(mapFunckijaPM25,axis=1)

def mapFunckijaPM10(row):
    row['AQI_PM10'] = train[row.name - datetime.timedelta(days=1) : row.name]['PM10'].mean()
    return row
    #return train['PM25'][datetime.datetime[]]
train = train.apply(mapFunckijaPM10,axis=1)

def mapFunckijaO3(row):
    row['AQI_O3'] = train[row.name - datetime.timedelta(days=1) : row.name]['O3'].mean()
    return row
    #return train['PM25'][datetime.datetime[]]
train = train.apply(mapFunckijaO3,axis=1)

def mapFunkcijaPM10history(row):
    row['PM10_history'] = train[row.name-datetime.timedelta(hours=3) : row.name-datetime.timedelta(hours=1)]['PM10'].mean()
    return row
train = train.apply(mapFunkcijaPM10history,axis=1)

print(train.columns)
target_cols = ['O3', 'PM10', 'PM25', 'cloudCover', 'dewPoint', 'humidity',
       'precipAccumulation', 'precipIntensity', 'precipProbability',
       'pressure', 'temperature', 'uvIndex', 'visibility', 'windBearing',
       'windSpeed','AQI_PM25', 'AQI_PM10', 'AQI_O3','PM10_history']
def mapFunkcijaDelta(row):
    for col in target_cols:
        row[col+"_delta"] = train[row.name:row.name][col] - train[row.name-datetime.timedelta(hours=1)
                                                                 :row.name-datetime.timedelta(hours=1)][col]
    return row

train.iloc[1:,:] = train.iloc[1:,:].apply(mapFunkcijaDelta,axis=1)
```

```python
#After clean
train["PM10_history"].interpolate(inplace=True, limit=3,method='time',limit_direction='both') # adapt limit here
train['PM10_history'].fillna(train['PM10_history'].mean(),inplace=True)
total = train.isnull().sum().sort_values(ascending=False)
percent = (train.isnull().sum()/train.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data = missing_data[missing_data['Total']>0]
missing_data
```

```python
# Skew analysis here
# SKEW ANALYSIS

numeric_feats = train.dtypes[train.dtypes != "object"].index

# Check the skew of all numerical features
skewed_feats = train[numeric_feats].apply(lambda x: skew(x.dropna())).sort_values(ascending=False)
print("\nSkew in numerical features: \n")
skewness = pd.DataFrame({'Skew' :skewed_feats})
skewness.head(30)
```

```python
# BOX COT TRANSFORMATION OR LOG TRANSFORMATION OF SKEWED FEATURES
skewness = skewness[(skewness.Skew) > 0.75] #define thresh here, cv later
print("There are {} skewed numerical features to Box Cox transform".format(skewness.shape[0]))

skewed_features = skewness.index
lam = 0.15
for feat in skewed_features:
    #all_data[feat] += 1
    train[feat] = boxcox1p(train[feat], lam)
```
